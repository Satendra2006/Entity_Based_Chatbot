```python
def predict_intent_with_confidence(text):
    text = clean_text(text)
    vec = vectorizer.transform([text])
    
    scores = model.decision_function(vec)
    confidence = max(scores[0])
    
    intent = model.predict(vec)[0]
    
    if confidence < 0.5:
        return "unknown", confidence
    
    return intent, confidence
```

```python
print(predict_intent_with_confidence("hi"))
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
def extract_entity(text):
    doc = nlp(text.lower())
    
    # collect meaningful noun chunks
    entities = []
    
    for chunk in doc.noun_chunks:
        # remove stop words like "the", "is"
        clean_chunk = " ".join([
            token.text for token in chunk 
            if not token.is_stop
        ])
        
        if clean_chunk:
            entities.append(clean_chunk)
    
    # return most meaningful (longest phrase usually best)
    if entities:
        return max(entities, key=len)
    
    return None
```

```python
tests = [
    "where is ece lab",
    "hostel fee details",
    "placement statistics",
    "library timing",
    "who is cse hod"
]

for t in tests:
    print(t, "→", extract_entity(t))
```

```python
def clean_entity(entity):
    if entity:
        words_to_remove = ["the", "is", "of","to"]
        words = entity.split()
        words = [w for w in words if w not in words_to_remove]
        return " ".join(words)
    return None
```

```python
text = "Where is the ECE lab?"
entity = clean_entity(extract_entity(text))

print(entity)
```

# Combing the Intent model and entity extraction model

```python
def fallback_entity(intent):
    fallback_map = {
        "timing": "college timing",
        "placement": "placement details",
        "fees": "fee structure",
        "admission":"admission process",
    }
    return fallback_map.get(intent, None)
```

```python
def predict(text):
    intent = predict_intent(text)
    entity = extract_entity(text)
    
    if entity is None:
        entity = fallback_entity(intent)
    
    return {
        "intent": intent,
        "entity": entity
    }
```

```python
print(predict("what are admission steps"))
```

```python
tests = [
    "where is cse block",
    "where is ece lab",
    "library timing",
    "what is college timing",
    "hostel facilities",
    "hostel fees",
    "placement statistics",
    "average package",
    "how to apply for admission",
    "who is cse hod"
]
```

```python
for t in tests:
    intent = predict_intent(t)
    entity = clean_entity(extract_entity(t))
    
    print(f"Input: {t}")
    print(f"Intent: {intent}, Entity: {entity}")
    print("-"*50)
```
